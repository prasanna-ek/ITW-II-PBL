```python
real_madrid=fifa[fifa['club']=='Real Madrid']
```

```python

names = list(real_madrid['short_name'][0:5])
wages = list(real_madrid['wage_eur'][0:5])
plt.figure(figsize=(10,10))


def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct

plt.pie(wages, labels=names, autopct=make_autopct(wages))
plt.show()
```

```python
fifa1 = fifa[(fifa['short_name']=='L. Messi') | (fifa['short_name']=='Cristiano Ronaldo') ]
fifa1.head()
```

```python

messivronalodo=fifa1[['short_name','pace','shooting','passing','dribbling','defending','physic','attacking_crossing','attacking_finishing','attacking_heading_accuracy','attacking_short_passing','attacking_volleys','skill_dribbling','skill_curve','skill_fk_accuracy','skill_long_passing','skill_ball_control','movement_acceleration','movement_sprint_speed','power_strength','power_jumping']]
```

```python
messivronalodo.head()
```

```python
params = list(messivronalodo.columns)
params = params[1:7]
params
```

```python
ranges = []
a_value = []
b_value = []

for x in params:
  a = min(messivronalodo[params][x])
  a = a - (a*.10)

  b = max(messivronalodo[params][x])
  b = b + (b*.10)

  ranges.append((a,b))

for x in range(len(messivronalodo['short_name'])):
  if messivronalodo['short_name'][x] == 'L. Messi':
    a_value = messivronalodo.iloc[x].values.tolist()
  if messivronalodo['short_name'][x] == 'Cristiano Ronaldo':
    b_value = messivronalodo.iloc[x].values.tolist()
```

```python
ranges=ranges[0:6]
```

```python
a_value=a_value[1:7]
b_value=b_value[1:7]

values=[a_value,b_value]
```

```python
values
```

```python
title = dict(
    title_name = 'L. Messi',
    title_color = 'blue',
    subtitle_name = 'Argentina',
    subtitle_color = 'blue',
    title_name_2 = 'Cristiano Ronaldo',
    title_color_2 = 'red',
    subtitle_name_2 = 'Portugal',
    subtitle_color_2 = 'red',
    title_fontsize=18,
    subtitle_fontsize=15
)
```

```python
n_groups = 6
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8

rects1 = plt.bar(index, a_value, bar_width,
alpha=opacity,
color='b',
label='L. messi')

rects2 = plt.bar(index + bar_width, b_value, bar_width,
alpha=opacity,
color='g',
label='C. Ronaldo')

plt.xlabel('Player')
plt.ylabel('Rating')
plt.title('Scores by person')
plt.xticks(index + bar_width, params)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
radar = Radar()

fig,ax = radar.plot_radar(ranges=ranges,params=params,values=values,radar_color=['blue','red'],alphas=[.75,.6],title=title,compare=True)
```
